# Datatypes:
Starting off with the basic python tools we went over last meeting.

Variables are used to store data. They can be of different types, such as integers, floats, strings, etc.

```python
# Integer
x = 5 # You can assign a variable using the = sign
print(x) # Prints variables to the console
```

```python
# Float
y = 5.0
# They are called floats because they can store decimal values, and thus the decimal point "floats" around.
print(y)
```

```python
# String
z = "Hello, "
# They are called strings because they are a "string" of characters.

# You can also use single quotes for strings.
z2 = 'world!'
print(z + z2) # You can concatenate strings using the + operator.
```

```python
# Boolean
a = True
b = False
print("a: " + str(a), "b: " + str(b)) # You can print multiple variables at once by separating them with commas.
# Only strings can be concatenated with the + operator, so we use the str() function to convert the boolean to a string.

# Booleans are used to store True or False values.
# You can operate on them using logical operators such as and, or, not, etc.
c = a and b
d = a or b
e = not a
print("a and b:", c)
print("a or b:", d)
print("not a:", e)
```

```python
# Lists
f = [1, 2, 3, 4, 5]
# Lists are used to store multiple values in a single variable.
# They are ordered, changeable, and allow duplicate values.
print("f:", f)

# Some operations you can do with lists:
print("First element:", f[0]) # Access the first element of the list
print("Last element:", f[-1]) # Access the last element of the list
print("Range from 1 to 3:", f[1:3]) # Access a range of elements in the list
print("Length:", len(f)) # Get the number of elements in the list

f.append(6) # Add an element to the end of the list
print("Appended 6 to end:", f)

f.insert(0, 0) # Add an element at a specific index
print("Inserted 0 at index 0:", f)

f.pop() # Remove the last element from the list
print("Popped last element:", f)

f.remove(0) # Remove the first element from the list
print("Removed first element:", f)
```

```python
# Tuple
g = (1, 2, 3, 4, 5)
# Tuples are used to store multiple values in a single variable.
# They are ordered, unchangeable, and allow duplicate values.
print("g:", g)

# Some operations you can do with tuples:
print("First element:", g[0]) # Access the first element of the tuple
print("Last element:", g[-1]) # Access the last element of the tuple
print("Range from 1 to 3:", g[1:3]) # Access a range of elements in the tuple
print("Length:", len(g)) # Get the number of elements in the tuple

# This will throw an exception because tuples are immutable and you cannot append elements to them.
try: # Try to do something
    print(g.append(6)) 
except Exception as e: # If an error occurs, catch it and store it in the variable e
    print("Exception:", e) 

# The main difference between lists and tuples is that lists are mutable (can be changed) and tuples are immutable (cannot be changed).
# This means that appending, inserting, removing, etc. elements from a tuple will throw an exception.

# You can convert a list to a tuple and vice versa using the list() and tuple() functions.
h = list(g)
print("Converted tuple to list:", h)
i = tuple(f)
print("Converted list to tuple:", i)
```

```python
# Set
j = {1, 2, 3, 4, 5}
# Sets are used to store multiple values in a single variable.
# They are unordered, unchangeable, and do not allow duplicate values.
print("j:", j)

# Some operations you can do with sets:
print("Is 1 in j?", 1 in j) # Check if an element is in the set

j.add(6) # Add an element to the set
print("Added 6 to set:", j)

j.remove(6) # Remove an element from the set
print("Removed 6 from set:", j)

print("Length:", len(j)) # Get the number of elements in the set
```

```python
# Dictionary
k = {
    "name": "John",
    "age": 30,
    "city": "New York"
}

# Dictionaries are used to store data in key-value pairs.
# They are unordered, changeable, and do not allow duplicate keys.
print(k)

# Some operations you can do with dictionaries:
print("What is your name?", k["name"]) # Access the value of a key
print("What is your age?", k["age"])
print(k.keys()) # Get the keys of the dictionary
print(k.values()) # Get the values of the dictionary
print(k.items()) # Get the key-value pairs of the dictionary

k["age"] = 40 # Change the value of a key
print("Changed age to 40:", k)
```

# Loops

Loops repeat sections of code (sometimes depending on a Boolean condition).

```python
# While loop
k = 0 # Technically reusing the variable k from earlier, but it's fine because it's overwritten
print("Printing numbers from 0 to 4:")
while k < 5: # While a condition is true, execute the code block
    print(k)
    k += 1 # Increment the value of k by 1
```

```python
# For loops
# These are used to iterate over a sequence (list, tuple, set, dictionary, etc.)
# You will almost always be using for loops instead of while loops in practice.
l = [1, 2, 3, 4, 5]
print("Printing elements in list:")
for m in l: # For each element in the list l, execute the code block
    print(m)

print("Printing numbers from 0 to 4:")
for i in range(5): # For each number from 0 to 4 (end is exclusive), execute the code block
    print(i)

print("Printing numbers from 1 to 10, stepping by 3:")
for i in range(1, 10, 3): # For each number from 1 to 10, incrementing by 3, execute the code block
    print(i)

print("Printing numbers from 1 to 2, incrementing by 0.1:")
try:
    for i in range(1, 2, 0.1): # You cannot use a float as a step size, so this will throw an exception
        print(i)
except Exception as e:
    print("Exception:", e)

print("Printing numbers from 5 to 1:")
for i in range(5, 0, -1): # You can also go backwards by specifying a negative step size
    print(i)

d = {
    "name": "John",
    "age": 30,
    "city": "New York"
}

print("Printing list with index:")
for i, m in enumerate(l): # For each index and element in the list l, execute the code block
    print(i, m) # Indexing starts from 0, so the first element has an index of 0

print("Printing keys and values in dictionary:")
for key, value in d.items(): # For each key-value pair in the dictionary d, execute the code block
    print(key, value)

# You can also use list comprehensions to create lists in a more concise way.
n = [i for i in range(5)] # Create a list of numbers from 0 to 4
print("List comprehension:", n)

# You can also use conditional statements in list comprehensions.
o = [i for i in range(10) if i % 2 == 0] # Create a list of even numbers from 0 to 9
print("List comprehension of evens:", o)
```

# Functions
Functions (or procedures if you're weird) are typically used to segment code into more understandable chunks.
It also provides a comprehensible abstraction for using code made by other people, like with those imports at the start.

```python
def add(a, b): # Define a function called add that takes two arguments a and b
    return a + b # Return the sum of a and b

# Not very useful, but you get the idea.
print("1 + 2 =", add(1, 2))

# You can also define functions with default arguments.
def decrement(a, b=1): # Define a function called decrement that takes two arguments a and b, with b having a default value of 1
    return a - b

print("5 - 1 =", decrement(5)) # If you don't specify a value for b, it will use the default value of 1
print("5 - 2 =", decrement(5, 2)) # You can also specify a value for b

# You can also define functions that take a variable number of arguments.
def add_all(*args): # Define a function called add_all that takes any number of arguments
    return sum(args) # Return the sum of all the arguments

print("1 + 2 + 3 =", add_all(1, 2, 3)) # You can pass any number of arguments to the function
print("1 + 2 + 3 + 4 =", add_all(1, 2, 3, 4))

# You can also define functions that take a variable number of keyword arguments.
def print_all(**kwargs): # Define a function called print_all that takes any number of keyword arguments
    for key, value in kwargs.items(): # For each key-value pair in the keyword arguments, execute the code block
        print(key, value)
print_all(name="John", age=30, city="New York") # You can pass any number of keyword arguments to the function

# Now, let's show some more advanced examples of functions.
# Remember, anything you can do with functions, you can also do without them (but it's much more of a pain).

def fizzbuzz(n): # Define a function called fizzbuzz that takes one argument n
    for i in range(1, n + 1): # For each number from 1 to n, execute the code block
        if i % 3 == 0 and i % 5 == 0: # If the number is divisible by 3 and 5, print "FizzBuzz"
            print("FizzBuzz")
        elif i % 3 == 0: # If the number is divisible by 3, print "Fizz"
            print("Fizz")
        elif i % 5 == 0: # If the number is divisible by 5, print "Buzz"
            print("Buzz")
        else: # Otherwise, print the number
            print(i)

print("fizzbuzz(15):")
fizzbuzz(15) # Call the fizzbuzz function with n = 15
```

```python
import numpy as np

# Now, let's show an example of using NumPy to perform some basic numerical operations.
# Generate a random matrix of shape (3, 3)
matrix = np.random.randn(3, 3)
print("Matrix:")
print(matrix)

# Transpose the matrix (flip the rows and columns)
transposed = matrix.T
print("Transposed Matrix:")
print(transposed)

# Invert the matrix
inverted = np.linalg.inv(matrix)
print("Inverted Matrix (when multiplied by the original matrix, it should give the identity matrix):")
print(inverted)

# Multiply the matrix by its inverse to get the identity matrix
identity = np.dot(matrix, inverted)
print("Identity Matrix (when multiplied by any matrix, it gives the same matrix):")
print(identity)

# Calculate the determinant of the matrix (scale factor on area/volume)
determinant = np.linalg.det(matrix)
print("Determinant of Matrix:")
print(determinant)

# Calculate the eigenvalues and eigenvectors of the matrix
eigenvalues, eigenvectors = np.linalg.eig(matrix)
print("Eigenvalues:") # Scale factor on eigenvectors
print(eigenvalues)
print("Eigenvectors:") # General "direction" of transformation
print(eigenvectors)
```

```python
# Let's visualize some basic transformations using NumPy and Seaborn.
from matplotlib.animation import FuncAnimation 

import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# Set Seaborn style
sns.set()

# Create a grid of points from -10 to 10
x_vals = np.arange(-10, 11)
y_vals = np.arange(-10, 11)
X, Y = np.meshgrid(x_vals, y_vals)
grid_points = np.vstack([X.ravel(), Y.ravel()])  # Shape is (2, N)

# Define transformation matrices
identity_matrix = np.array([[1, 0],
                            [0, 1]])

# Scale matrix (scales x by 2 and y by 0.5)
scale_matrix = np.array([[2, 0],
                         [0, 0.5]])

# Rotation matrix (rotates by 45 degrees)
theta = np.radians(45)
rotation_matrix = np.array([[np.cos(theta), -np.sin(theta)],
                            [np.sin(theta),  np.cos(theta)]])

# Shear matrix (shears in the x-direction by 0.5)
shear_matrix = np.array([[1, 0.5],
                         [0, 1]])

# Scale and shear combination matrix
scale_shear_matrix = np.array([[1.5, 0.5],
                               [0.2,   1]])

# Random transformation matrix
np.random.seed(0)  # For reproducibility
random_matrix = np.random.uniform(-1, 1, size=(2, 2))

# List of transformations
transformations = [
    ('Identity', identity_matrix),
    ('Scale', scale_matrix),
    ('Rotation', rotation_matrix),
    ('Shear', shear_matrix),
    ('Scale and Shear', scale_shear_matrix),
    ('Random', random_matrix)
]

# Function to animate transformations
def animate_transformation(start_matrix, end_matrix, grid_points, title, frames=40, interval=50):
    # Generate interpolation values
    t_values = np.concatenate((np.linspace(0, 1, frames // 2),
                               np.linspace(1, 0, frames // 2)))
    transformed_data = []
    for t in t_values:
        # Interpolate between start and end matrices
        interp_matrix = (1 - t) * start_matrix + t * end_matrix
        # Apply transformation
        transformed_points = np.dot(interp_matrix, grid_points)
        transformed_data.append(transformed_points)
    
    fig, ax = plt.subplots(figsize=(6, 6))
    # Initialize scatter plot
    sc = ax.scatter(transformed_data[0][0], transformed_data[0][1], s=10, color='blue')
    
    # Set plot limits and properties
    ax.set_xlim(-20, 20)
    ax.set_ylim(-20, 20)
    ax.set_aspect('equal')
    ax.set_title(title)
    
    # Update function for animation
    def update(frame):
        sc.set_offsets(transformed_data[frame].T)
        return sc,

    # Create animation
    anim = FuncAnimation(fig, update, frames=len(transformed_data), interval=interval, blit=True)
    plt.close(fig)  # Prevents duplicate display of static plots
    return anim

# Create and display animations for each transformation
animations = []
for name, matrix in transformations:
    anim = animate_transformation(identity_matrix, matrix, grid_points, title=name)
    animations.append((name, anim))

# Display animations (this code depends on the environment)
from IPython.display import HTML, display

for name, anim in animations:
    display(HTML(f'<h3>{name} Transformation</h3>'))
    display(HTML(anim.to_jshtml()))

# The main point is to show you some of the cool things you can do with NumPy and Seaborn.
# If you are interested in learning more, I encourage you to explore the documentation and experiment with the code.
# Making matplotlib animations is a bit more advanced, so don't worry if you don't understand everything.
```

# ANNs:
Artificial Neural Networks (ANNs) are one of the most exciting areas in technology today. They allow computers to learn from data and make decisions or predictions. This guide will help you understand what neural networks are, how they work, and how they're changing the world—all without needing deep mathematical knowledge.

# The Biological Inspiration:
Neural networks are inspired by the human brain.

**Neurons and Synapses in the Brain**
- **Neurons**: The basic units of the brain and nervous system. They receive input, process it, and send output to other neurons.
- **Synapses**: Connections between neurons that allow them to communicate.
*In the brain, learning happens by strengthening or weakening these connections based on experiences.*

An Artificial Neural Network is a computational model that mimics the way biological neurons signal to one another.
- **Purpose**: To recognize patterns and relationships in data through a learning process.
- **Functionality**: They can learn from data without being explicitly programmed with task-specific rules.

While we get an inspiration from the brain, in reality all models do is use data to draw "lines".
These "lines" group together areas, which allows for interpolation, categorizing new data.

# Structure:
- **Artificial Neurons**: Also called nodes; they receive input, process it, and produce output.
- **Processing**: Each neuron sums its inputs and passes the result through an activation function.
    - An activation function is used after every node does its magic so that it can fit to fancier data.
- **Layers**
    - **Input Layer**: Where the network receives input data.
    - **Hidden Layers**: Layers between input and output layers where computations occur. This can be thought of as "pattern recognition".
    - **Output Layer**: Where you take the output from the Neural Network.

When considering a model that takes in "images" a.k.a. a vision model, you can think of each hidden layer as categorizing "patterns".
If you're categorizing drawn numbers, you can think of loops, lines, strokes, with the further hidden layers finding patterns of these patterns.

# Weights and Biases
- **Weights**: Values that modify the input signals. They represent the strength of connections between neurons.
- **Biases**: Additional values added to the weighted sum to adjust the output.

Now, how do we actually train them?
Calculus. But most of you won't learn that for a bit. So I'm gonna talk in aphormisms. Kind of.

**Forward propagation** is the process by which input data moves through the neural network to generate an output. It's like a flow of information from the input layer, through any hidden layers, to the output layer.

Think of a neural network as a waterfall. The water (input data) flows from the top (input layer), cascades over rocks and obstacles (hidden layers and neurons), and eventually reaches the bottom (output layer). At each stage, the water is guided and redirected, similar to how data is transformed by weights and activation functions in each neuron.

**Backpropagation** is the method by which the neural network learns from its mistakes. It adjusts the weights and biases of the network to minimize the difference between its predicted output and the actual output.

Imagine you're learning to play the piano by ear. You press a key hoping to play a specific note. If the sound isn't right, you listen to the difference between what you played and what you intended, then adjust your finger placement accordingly. 

Backpropagation involves the network adjusting its parameters based on the error of its output compared to the desired output.
To handwave the underlying calculus, you can think of it as seeing how changing each value changes how much you suck.
If moving a value in one way makes the model suck less, it'll move in that direction.
This just happens with like a bajillion parameters.

To enumerate the process:
- **Calculate Error**: After forward propagation, determine the difference between the predicted output and the actual output using a loss function.
- **Propagate Error Backwards**: Starting from the output layer, the error is sent back through the network.
- **Adjust Weights and Biases**: The network updates its weights and biases to reduce the error for the next forward pass.

Forward propagation is using the model, Backward propagation is it looking back at its mistakes and getting better.

```python
# Now to the fun part: machine learning!
# We will use PyTorch to create a simple neural network to fit to the linear function: y = 3x + 1.

import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt

torch.manual_seed(0)  # For reproducibility, random numbers generated by PyTorch will be the same every time

# Generate some random data
x_vals = torch.linspace(-10, 10, 100).view(-1, 1)  # Shape is (100, 1)
# Apply linear function y = 3x + 1 with some noise
y_vals = 3 * x_vals + 1 + torch.randn_like(x_vals) * 2  # Shape is (100, 1)

"""
torch.linspace: Generates 100 evenly spaced values between -10 and 10.
torch.view: Reshapes the tensor to have one column (-1 infers 100 rows).
torch.randn_like: Generates random noise with the same shape as x_vals. Multiplying by 2 makes the standard deviation 2.
y_vals = 3 * x_vals + 1 + noise: Applies the linear function with some noise to simulate real-world imperfect data.
"""

# Let's visualize the data.
plt.scatter(x_vals, y_vals, label='Data', color='blue')
```

```python
# Neat, isn't it?
# Now, let's create a simple neural network to fit the data.
# We will use a single linear layer with one input and one output. I'll explain more about the structure of the network later.

# Define the neural network
class LinearRegression(nn.Module):
    def __init__(self):
        super(LinearRegression, self).__init__()
        self.linear = nn.Linear(1, 1)  # Create a linear layer with one input and one output

    def forward(self, x):
        out = self.linear(x)
        return out
# I technically didn't need to create a class for this simple model, but it's good practice for more complex models.
# A class is a blueprint for creating objects, and an object is an instance of a class. You'll learn more about this in Java.

# Make an instance of the model
model = LinearRegression()

# nn.Module: Base class for all neural network modules in PyTorch.
# nn.Linear(1, 1): Creates a linear transformation ( y = Wx + b ) with one input feature and one output feature.
# forward(self, x): Defines the forward pass of the model.
```

```python
# Define the loss function and optimizer
criterion = nn.MSELoss()  # Mean squared error loss
learning_rate = 0.01
optimizer = optim.SGD(model.parameters(), lr=learning_rate)  # Stochastic gradient descent optimizer

# nn.MSELoss(): Measures the mean squared error between the model's predictions and the targets.

# MSE = 1/n * Σ (y - y')²
# , which is the average of the squared differences between the predictions and the targets.

# learning_rate: Determines the step size at each iteration while moving toward a minimum of a loss function.
# optim.SGD: Implements the stochastic gradient descent optimizer. We pass the model parameters and the learning rate.
```

```python

# Train the model
num_epochs = 100
loss_values = [] # Store the loss values at each epoch

for epoch in range(num_epochs):
    # Zero gradients
    optimizer.zero_grad()
    # Forward pass
    outputs = model(x_vals)
    # Calculate loss
    loss = criterion(outputs, y_vals)
    # Backward pass (compute gradients)
    loss.backward()
    # Update weights
    optimizer.step()
    # Store loss value
    loss_values.append(loss.item())

    # Print progress every 10 epochs
    if (epoch + 1) % 10 == 0:
        print(f'Epoch [{epoch + 1}/{num_epochs}], Loss: {loss.item():.4f}')

# optimizer.zero_grad(): Clears old gradients. Gradients are accumulated by default in PyTorch.
# outputs = model(x_values): Computes the model's predictions.
# loss = criterion(outputs, y_values): Computes the loss between predictions and actual values.
# loss.backward(): Computes the gradient of the loss w.r.t. model parameters.
# optimizer.step(): Updates the model parameters using the computed gradients.
```
